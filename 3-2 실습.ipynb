```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
value = np.random.standard_normal(40)
value
```

```python
index = range(len(value))
plt.plot(index, value)

plt.xlim(0, 20)
plt.ylim(np.min(value) -0.5, np.max(value) + 0.5)
```

```python
plt.figure(figsize=(7,4))
plt.plot(value.cumsum(), 'b', lw= 1.5)
plt.plot(value.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 1')
```

```python
value = np.random.standard_normal((30, 2))
value
```

```python
plt.figure(figsize=(10,4))
plt.plot(value[:,0], lw= 1.5, label='1st')
plt.plot(value[:,1], lw= 1.5, label='2nd')
plt.plot(value, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 2')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(value[:,0], lw = 1.5, label = '1st')
plt.plot(value[:,0], 'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.ylabel("value")
plt.title('Line Plot 3')
```

```python
plt.subplot(212)
plt.plot(value[:,1], 'g', lw = 1.5, label = '2nd')
#plt.plot(value[:,1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
```

```python
value = np.random.standard_normal((100,2))
plt.plot(value[:,0], value[:,1], 'ro')
plt.grid(False)
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 1')
```

```python
plt.figure(figsize=(7,5))
plt.scatter(value[:,0], value[:,1], marker='o')
plt.grid(True)
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 2')
```

```python
color = np.random.randint(0,10,len(value))

plt.figure(figsize=(10,5))
plt.scatter(value[:,0], value[:,1], c=color, marker = 'o')
plt.colorbar()
plt.xlabel('value 1')
plt.ylabel('value 2')
plt.title('Scatter Plot 3')
```
