```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
value = np.random.standard_normal((500,2))
plt.figure(figsize=(12,7))
plt.hist(value, label = ['1st', '2nd'], bins = 30)
plt.grid(True)
plt.legend(bbox_to_anchor = (1.1, 1), loc="upper right")
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 1')
```

```python
plt.figure(figsize=(12,6))
plt.hist(value, label = ['1st', '2nd'], bins = 30, color=('pink', 'green')) #핑크예쁘다
plt.grid(True)
plt.legend(bbox_to_anchor = (1.1, 1), loc="upper right")
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 2')
```

```python
value = np.random.standard_normal(500)
cm = plt.cm.get_cmap('Spectral')
plt.figure(figsize=(12,6))

n, bins, patches = plt.hist(value,bins = 30, color='yellow')
bin_centers = 0.5 * (bins[:-1] + bins[1:])

col = bin_centers - min(bin_centers)
col = col / max(col)

for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))

plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 3')
plt.show()
```

```python
pivoted_data = data2.pivot("year", "month", "passengers")
pivoted_data
```

```python
sns.heatmap(pivoted_data, annot=True, fmt= "d", linewidth=3)
```
